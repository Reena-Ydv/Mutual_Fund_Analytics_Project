```python
import pandas as pd
import numpy as np

# Load Dataset
fdf = pd.read_csv("../Data/Raw/01_fund_master.csv")

# Display basic information
print("Shape:", fdf.shape)
display(fdf.head())

fdf.info()

print("\nMissing Values:")
print(fdf.isnull().sum())

print(fdf.columns.tolist())
```

```python
# Converting launch_date to datetime
fdf["launch_date"] = pd.to_datetime(
    fdf["launch_date"],
    errors="coerce"
)

print(fdf["launch_date"].isna().sum())
```

```python
# Checking Duplicate AMFI Codes
print("Duplicate AMFI Codes:", fdf["amfi_code"].duplicated().sum())
```

```python
# Validating Expense Ratio
invalid_expense = fdf[
    (fdf["expense_ratio_pct"] < 0.1) |
    (fdf["expense_ratio_pct"] > 2.5)
]

print("Invalid Expense Ratio Records:", len(invalid_expense))
```

```python
# Validating Minimum SIP Amounts
print("Invalid SIP Amounts:",
      (fdf["min_sip_amount"] <= 0).sum())
```

```python
# Validating Minimum Lumpsum Amount

print("Invalid Lumpsum Amounts:",
      (fdf["min_lumpsum_amount"] <= 0).sum())
```

```python
# Validating Exit loads
print("Negative Exit Load:",
      (fdf["exit_load_pct"] < 0).sum())
```

```python
fdf.to_csv(
    "../Data/Processed/01_fund_master_cleaned.csv",
    index=False
)

print("01_fund_master_cleaned.csv saved successfully!")
```
