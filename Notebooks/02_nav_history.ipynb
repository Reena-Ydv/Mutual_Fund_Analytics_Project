```python
nav_df = pd.read_csv("../Data/Raw/02_nav_history.csv")
nav_df.head()
```

```python
print(nav_df.columns.to_list())
print(nav_df.shape)
```

```python
nav_df['date']= pd.to_datetime(nav_df['date'],errors='coerce')
# converts date column to datetime format, errors='coerce' will convert invalid parsing (date) to NaT
```

```python
nav_df.info()
```

```python
nav_df= nav_df.sort_values(by=['amfi_code','date']).reset_index(drop=True)
nav_df.head()
```

```python
nav_df.isna().sum()
```

```python
nav_df.info()
```

```python
nav_df["nav"].isnull().sum()
```

```python
nav_df[nav_df["nav"].isnull()]
```

```python
# checking for duplicate rows in the dataframe
duplicate_count = nav_df.duplicated().sum()
print(f"Duplicate rows: {duplicate_count}")
```

```python
nav_df = nav_df[nav_df["nav"] > 0].reset_index(drop=True)
```

```python
# checking processed dataframe information
print(nav_df.info())
print(nav_df.describe())
```

```python
# saving the cleaned dataframe to a new CSV file
nav_df.to_csv("../Data/Processed/02_nav_history_cleaned.csv",index=False)
```
