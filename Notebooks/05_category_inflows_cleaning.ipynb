# 05 Category Flows - Data Cleaning

## Objective

Clean and preprocess the category-wise mutual fund flows dataset before using it for Exploratory Data Analysis (EDA).

The cleaning process includes:

- Loading the raw dataset
- Inspecting dataset structure
- Identifying missing values
- Checking duplicate records
- Correcting data types
- Validating cleaned data
- Exporting the cleaned dataset

```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
category_df = pd.read_csv("../Data/Raw/05_category_inflows.csv")

print("DATASET INFORMATION")


display(category_df.head())

print("\nShape:", category_df.shape)

print("\nColumns:")
print(category_df.columns)

print("\nData Types:")
print(category_df.dtypes)

print("\nMissing Values:")
display(category_df.isnull().sum())

print("\nDuplicate Records:")
print(category_df.duplicated().sum())
```

```python
# Data Cleaning

# Convert month to datetime
category_df["month"] = pd.to_datetime(category_df["month"])

# Sort chronologically
category_df = category_df.sort_values(["month", "category"]).reset_index(drop=True)

# Verify
category_df.info()

category_df.head()
```

```python
print("DATA VALIDATION")


print(f"Shape : {category_df.shape}")
print(f"Duplicate Rows : {category_df.duplicated().sum()}")

print("\nMissing Values")
print(category_df.isnull().sum())

print("\nDate Range")
print(category_df["month"].min())
print(category_df["month"].max())

print("\nUnique Categories :", category_df["category"].nunique())
```

```python
output_path = "../Data/Processed/05_category_inflows_cleaned.csv"

category_df.to_csv(output_path, index=False)

print(f"Cleaned dataset saved to:\n{output_path}")
```

## Cleaning Summary

### Cleaning Performed

- Converted the `month` column from object to datetime format.
- Sorted the dataset by month and category.
- Verified duplicate records.
- Verified missing values.
- Exported the cleaned dataset to the Processed folder.

### Observations

- No missing values were identified.
- No duplicate records were present.
- The dataset required only a datetime conversion for the `month` column.

- Records: 144
- Features: 3
- Date Range: April 2024 – March 2025
