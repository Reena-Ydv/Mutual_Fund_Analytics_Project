# 04 SIP Monthly Inflows - Data Cleaning

## Objective

Clean and preprocess the monthly SIP inflow dataset before using it for Exploratory Data Analysis (EDA). The cleaning process includes:

- Loading the raw dataset
- Inspecting data types and structure
- Checking missing values
- Identifying duplicate records
- Standardizing data types
- Validating cleaned data
- Exporting the cleaned dataset to the Processed folder

```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
# Load raw dataset

sip_df = pd.read_csv("../Data/Raw/04_monthly_sip_inflows.csv")

sip_df.head()
```

```python
print("DATASET INFORMATION")

display(sip_df.head())

print("\nShape:", sip_df.shape)

print("\nColumns:")
print(sip_df.columns)

print("\nData Types:")
print(sip_df.dtypes)

print("\nMissing Values:")
display(sip_df.isnull().sum())

print("\nDuplicate Records:")
print(sip_df.duplicated().sum())
```

```python
# Convert month to datetime
sip_df["month"] = pd.to_datetime(sip_df["month"])

# Sort chronologically
sip_df = sip_df.sort_values("month").reset_index(drop=True)

# Verify
sip_df.info()
sip_df.head()
```

```python
print("DATA VALIDATION")

print(f"Shape : {sip_df.shape}")
print(f"Duplicate Rows : {sip_df.duplicated().sum()}")

print("\nMissing Values")
print(sip_df.isnull().sum())

print("\nDate Range")
print(sip_df["month"].min())
print(sip_df["month"].max())
```

```python
# Save cleaned dataset

output_path = "../Data/Processed/04_monthly_sip_inflows_cleaned.csv"

sip_df.to_csv(output_path, index=False)

print(f"Cleaned dataset saved to:\n{output_path}")
```

## Cleaning Summary

### Cleaning Performed

- Converted the `month` column to datetime format.
- Sorted the dataset chronologically.
- Verified that no duplicate records exist.
- Verified there are no missing values except for `yoy_growth_pct`.

### Observations

- The first 12 months (January–December 2022) have missing values in `yoy_growth_pct`.
- These missing values are expected because year-over-year growth cannot be calculated without data from the previous year.
- No imputation was performed to preserve data integrity.

### Final Dataset

- Records: 48
- Features: 6
- Date Range: January 2022 – December 2025
