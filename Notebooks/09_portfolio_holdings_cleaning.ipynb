# 09 Portfolio Holdings - Data Cleaning

## Objective

Clean and preprocess the Portfolio Holdings dataset before using it for Exploratory Data Analysis (EDA).

The cleaning process includes:

- Loading the raw dataset
- Inspecting dataset structure
- Checking missing values
- Identifying duplicate records
- Correcting data types
- Validating cleaned data
- Exporting the cleaned dataset

```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
holding_df = pd.read_csv("../Data/Raw/09_portfolio_holdings.csv")

print("DATASET INFORMATION")

display(holding_df.head())

print("\nShape:", holding_df.shape)

print("\nColumns:")
print(holding_df.columns)

print("\nData Types:")
print(holding_df.dtypes)

print("\nMissing Values:")
display(holding_df.isnull().sum())

print("\nDuplicate Records:")
print(holding_df.duplicated().sum())
```

```python
# Convert portfolio date to datetime
holding_df["portfolio_date"] = pd.to_datetime(holding_df["portfolio_date"])

# Sort records
holding_df = holding_df.sort_values(
    ["amfi_code", "weight_pct"],
    ascending=[True, False]
).reset_index(drop=True)

# Verify
holding_df.info()
```

```python
print("DATA VALIDATION")

print(f"Shape : {holding_df.shape}")
print(f"Duplicate Rows : {holding_df.duplicated().sum()}")

print("\nMissing Values")
print(holding_df.isnull().sum())

print("\nPortfolio Date")
print(holding_df["portfolio_date"].min())
print(holding_df["portfolio_date"].max())

print("\nUnique Schemes :", holding_df["amfi_code"].nunique())

print("Unique Sectors :", holding_df["sector"].nunique())
```

```python
output_path = "../Data/Processed/09_portfolio_holdings_cleaned.csv"

holding_df.to_csv(output_path, index=False)

print(f"Cleaned dataset saved to:\n{output_path}")
```

## Cleaning Summary

### Cleaning Performed

- Converted the `portfolio_date` column from object to datetime format.
- Sorted records by AMFI code and portfolio weight.
- Verified duplicate records.
- Verified missing values.
- Exported the cleaned dataset to the Processed folder.

### Observations

- No missing values were identified.
- No duplicate records were present.
- Only the `portfolio_date` column required data type conversion.
