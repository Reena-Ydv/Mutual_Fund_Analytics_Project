# 06 Industry Folio Count - Data Cleaning

## Objective

Clean and preprocess the Industry Folio Count dataset before using it for Exploratory Data Analysis (EDA).

The cleaning process includes:

- Loading the raw dataset
- Inspecting dataset structure
- Checking missing values
- Identifying duplicate records
- Correcting data types
- Validating cleaned data
- Exporting the cleaned dataset

```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
folio_df = pd.read_csv("../Data/Raw/06_industry_folio_count.csv")

print("DATASET INFORMATION")

display(folio_df.head())

print("\nShape:", folio_df.shape)

print("\nColumns:")
print(folio_df.columns)

print("\nData Types:")
print(folio_df.dtypes)

print("\nMissing Values:")
display(folio_df.isnull().sum())

print("\nDuplicate Records:")
print(folio_df.duplicated().sum())
```

```python
# Convert month to datetime
folio_df["month"] = pd.to_datetime(folio_df["month"])

# Sort chronologically
folio_df = folio_df.sort_values("month").reset_index(drop=True)

# Verify
folio_df.info()

folio_df.head()
```

```python
print("DATA VALIDATION")

print(f"Shape : {folio_df.shape}")
print(f"Duplicate Rows : {folio_df.duplicated().sum()}")

print("\nMissing Values")
print(folio_df.isnull().sum())

print("\nDate Range")
print(folio_df["month"].min())
print(folio_df["month"].max())
```

```python
output_path = "../Data/Processed/06_industry_folio_count_cleaned.csv"

folio_df.to_csv(output_path, index=False)

print(f"Cleaned dataset saved to:\n{output_path}")
```

## Cleaning Summary

### Cleaning Performed

- Converted the `month` column from object to datetime format.
- Sorted the dataset chronologically.
- Verified duplicate records.
- Verified missing values.
- Exported the cleaned dataset to the Processed folder.

### Observations

- No missing values were identified.
- No duplicate records were present.
- Only the `month` column required data type conversion.

- Records: 21
- Features: 6
- Date Range: January 2022 – December 2025
