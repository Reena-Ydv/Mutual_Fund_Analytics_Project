```python
# Loading the dataset
import pandas as pd
import numpy as np

# Load Dataset
adf = pd.read_csv("../Data/Raw/03_aum_by_fund_house.csv")

# Basic Exploration
print("Shape:", adf.shape)

display(adf.head())

adf.info()

print("\nMissing Values:")
print(adf.isnull().sum())

print("\nColumns:")
print(adf.columns.tolist())
```

```python
# Converting 'date' column to datetime
adf["date"] = pd.to_datetime(
    adf["date"],
    errors="coerce"
)

print("Invalid Dates:", adf["date"].isna().sum())
```

```python
# Checking for duplicates based on 'date' and 'fund_house'
duplicates = adf.duplicated(
    subset=["date", "fund_house"]
).sum()

print("Duplicate Records:", duplicates)
```

```python
# Validating AUM values
print(
    "Invalid AUM Records:",
    (adf["aum_crore"] <= 0).sum()
)
```

```python
# Validating Number of Schemes
print(
    "Invalid Number of Schemes:",
    (adf["num_schemes"] <= 0).sum()
)
```

```python
# Checking for summary statistics of numerical columns
adf.describe()
```

```python
# Saving the cleaned dataset to a new CSV file in Processed folder
adf.to_csv(
    "../Data/Processed/03_aum_by_fund_house_cleaned.csv",
    index=False
)

print("03_aum_by_fund_house_cleaned.csv saved successfully!")
```
